2. Проверим, есть ли в данных пропуски

```python
# Составим словарь, ключами в котором будут названия признаков, а значениями - число пропусков для данного признака
dict_with_values_nan = {}
for column in data_train.columns:
    dict_with_values_nan[column] = data_train[column].isna().sum() 
dict_with_values_nan
```

Можно заметить, что в данных присутствуют признаки, значения которых почти всегда пропущены. Так из 1460 наблюдений для признака "PoolQC" пропущено 1453 значений, для признака "MiscFeature" пропущено 1406 значений, для "Alley" - 1369 значений, а для признака "Fence" - 1179 значений. Необходимо подробнее остановиться на изучении данных признаков. Исходя из описания каждого признака, можно заключить, что данные признаки можно удалить, т.к. для данных признаков пропущено очень много значений. 

3. Построим распределение значений для каждого признака

```python
data_train.describe()
```

```python
def get_distribution(data, name_data):
    for feature in data.columns:
        plt.figure(figsize = (10, 5))
        data[feature].hist(bins=50)
        plt.ylabel('Count')
        plt.xlabel(feature)
        plt.title(f'{feature} distribution for {name_data}')
        plt.show()
```

```python
get_distribution(data_train, 'train data')
```

Для некоторых признаков наблюдается нормальное распределение: LotFrontage, YearBuilt, 1stFlrSF, SalePrice.

Из построенных графиков можно явно заключить, что данные часто несбалансированы в рамках признака.

6. Построим корреляционную матрицу для данных

```python
def get_correlation_heatmap(data, label):
    corr_mat = data.corr(method='pearson')
    plt.figure(figsize = (30, 15))
    sns.heatmap(corr_mat,vmax=1,square=True,annot=True,cmap='cubehelix')
```

```python
get_correlation_heatmap(data_train, "Корелляционная матрица для тренировочных данных")
```

Построим попарные графики для двух пар с максимальной корреляцией, чтобы проверить наличие линейной зависимости признаков. 

```python
sns.pairplot(data_train[['YearBuilt', 'GarageYrBlt']])
```

```python
sns.pairplot(data_train[['OverallQual', 'SalePrice']])
```

Для двух пар, приведенных выше, действительно существует линейная зависимость, как и ожидалось. 

Для признаков, корреляция между которыми близка к нулю, необходимо проводить дополнительное исследование ввиду того, что корреляция показывает только линейную зависимость.

Таким образом, в работе был проведен разведочный анализ и выявлено следующее:
- В данных присутствуют признаки, значения которых не определены во многих наблюдениях. Такие признаки лучше удалить, т.к. они имеют слишком мало значений.
- В данных присутствуют признаки, которые коррелируют между собой. Следовательно, при высокой степени зависимости можно объединить несколько признаков в один более информативный или удалить такие признаки, оставив только один из них. 
- Присутствуют признаки, содержащие большое количество аномальных значений. Такие признаки нуждаются в дополнительном исследовании. 
- Для нескольких рассмотренных признаков распределение значений и соотношения между ними соответствуют ожидаемым, однако требуется дополнительное иссследование для всех признаков.
